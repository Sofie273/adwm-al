```python
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
```

```python
#Beispieldatensatz:
rng = np.random.RandomState(42)
X = 0.3 * rng.randn(100, 2)
X_train = np.r_[X + 2, X - 2]
X_test = 0.3 * rng.randn(20, 2)
X_outliers = rng.uniform(low=-4, high=4, size=(20, 2))

X = np.r_[X_train, X_test, X_outliers]

df = pd.DataFrame(X, columns=['Feature1', 'Feature2'])
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Feature1'], df['Feature2'], color='blue', label='Datenpunkte')
plt.title("Beispiel Isolation Forest")
plt.xlabel("Feature1")
plt.ylabel("Feature2")
plt.legend()
plt.show()
```

```python
model = IsolationForest(contamination=0.1, random_state=42)
#contamination: Schwellwert
model.fit(df[['Feature1', 'Feature2']])

# Vorhersagen treffen (1: normal, -1: Anomalie)
df['Anomaly'] = model.predict(df[['Feature1', 'Feature2']])

# Anomalie-Score (je positiver umso typischer, je negativer umso eher Anomalie)
df['Anomaly_Score'] = model.decision_function(df[['Feature1', 'Feature2']])

random_seed = 4
print(df.sample(n=15, random_state=random_seed))
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Feature1'], df['Feature2'], c=df['Anomaly'], cmap='coolwarm')
plt.title("Detektierte Anomalien mit IF")
plt.xlabel("Feature1")
plt.ylabel("Feature2")
plt.show()
```
